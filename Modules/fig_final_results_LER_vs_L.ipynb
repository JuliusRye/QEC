```python
import matplotlib.pyplot as plt
```

```python
code_distance = [3,5,7,9]

CSS = [1.74E-3, 2.89E-4, 1.07E-4, None]
XZZX = [4.08E-4, 2.89E-5, 2.61E-5, None]
XY = [9.93E-5, 2.76E-5, 1.57E-4, None]
C1 = [3.15E-5, 1.08E-5, 5.03E-5, None]

hill_climb = [1.14E-4, 4.63E-5, 6.29E-5, None]

RL = [None, None, None, None]

PD = [1.04E-4, 2.25E-5, 3.19E-5, 3.08E-5]
```

```python
plt.figure()
plt.plot(code_distance, CSS, ls = '-', marker='o', label='CSS')
plt.plot(code_distance, XZZX, ls = '-', marker='o', label='XZZX')
plt.plot(code_distance, XY, ls = '-', marker='o', label='XY')
plt.plot(code_distance, C1, ls = '-', marker='o', label='C1')
plt.plot(code_distance, hill_climb, ls = '--', marker='p', label='Hill Climb')
plt.plot(code_distance, RL, ls = '--', marker='p', label='RL')
plt.plot(code_distance, PD, ls = ':', marker='^', label='PD')
plt.yscale('log')
plt.xlabel('Code Distance')
plt.ylabel('Logical Error Rate')
plt.legend()
plt.show()
```
