```python
from pymatching import Matching

import numpy as np
```

```python
H = np.array(
    [[1,1,0],
     [0,1,1],
     [1,0,1]]
)
error_probability = np.array(
    [.1,.2,.5]
)
N = np.log(error_probability.mean())
W = np.log(error_probability) / N
# W = np.log((1-error_probability)/error_probability)
display(W)

matching = Matching(H, weights=W)
matching.draw()
```

```python
error = np.array([0,1,1])
syndrome = H @ error % 2
matching.decode(syndrome)
```

```python
syndromes = np.random.randint(0,2, size=(20,3))
syndromes[:,0] = syndromes[:,1:].sum(axis=1) % 2
matching.decode_batch(syndromes)
```
