```python
import matplotlib.pyplot as plt
import numpy as np
```

# Logical error of CDSC using the CNN-G decoder

```python
arr = np.array([
    [1.78E-3, 3.91E-4, 3.93E-4, 1.54E-4],
    [3.18E-4, 6.83E-5, 1.76E-4, 7.03E-5],
    [4.43E-4, 4.60E-5, 1.04E-3, 4.44E-4],
])
arr.T
```

```python
code_distance = [3,5,7]

CSS = arr.T[0]
XZZX = arr.T[1]
XY = arr.T[2]
C1 = arr.T[3]

method_rs = [1.13E-4, 6.17E-5, 1.11E-4]
method_hc = [1.05E-4, 5.99E-5, 5.07E-5]

method_rl = [8.20E-5, 4.90E-5, 9.50E-5]

method_tdl = [3.03E-4, 4.99E-5, 4.73E-5]
```

```python
plt.figure(figsize=(5,3))

# plt.xkcd()

plt.plot(code_distance, CSS, ls = ':', marker='o', label='CSS')
plt.plot(code_distance, XZZX, ls = ':', marker='o', label='XZZX')
plt.plot(code_distance, XY, ls = ':', marker='o', label='XY')
plt.plot(code_distance, C1, ls = ':', marker='o', label='C1')
plt.plot(code_distance, method_rs, ls = '-', marker='p', label='Random Search')
plt.plot(code_distance, method_hc, ls = '-', marker='p', label='Hill Climb')
plt.plot(code_distance, method_rl, ls = '-', marker='p', label='Reinforcement Learning')
plt.plot(code_distance, method_tdl, ls = '-', marker='p', label='Thermodynamic Limit')

plt.yscale('log')
plt.xlabel('Code Distance')
plt.ylabel('Logical Error Rate')
plt.legend(ncols=2, loc='upper center', bbox_to_anchor=(0.5, 1.45))
plt.xticks(code_distance)
plt.grid(which='major')
plt.grid(which='minor', lw=.2)

plt.savefig('figs/ler_vs_L-compare.png', dpi=300, bbox_inches='tight')
plt.savefig('figs/ler_vs_L-compare.pdf', dpi=300, bbox_inches='tight')
plt.savefig('figs/ler_vs_L-compare.eps', dpi=300, bbox_inches='tight')

plt.show()
```
