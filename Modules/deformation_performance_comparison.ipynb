```python
import jax.numpy as jnp
import matplotlib.pyplot as plt
```

```python
all_logical_error_rates_CNN = jnp.load("data/environmentNoiselessCNN_error_rates_p01_nu500_halve.npy")
all_logical_error_rates_PML = jnp.load("data/environmentPML_error_rates_p01_nu500_halve.npy")
```

```python
sorted_order = all_logical_error_rates_PML.argsort()
plt.figure(dpi=300, figsize=(5,3))
plt.plot(all_logical_error_rates_CNN[sorted_order], '.', ms=2, label="CNN scores")
plt.plot(all_logical_error_rates_PML[sorted_order], '.', ms=2, label="PML scores")
plt.yscale('log')
plt.xlabel("Clifford deformed surface codes (Sorted by PML scores)")
plt.ylabel("Logical error rate")
plt.xlim(-100,20000)
plt.ylim(1E-5,1E-2)
plt.legend()
plt.show()
```
