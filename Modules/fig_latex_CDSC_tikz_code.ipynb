```python
from core.quantum_error_correction_code import SurfaceCode

import jax.numpy as jnp
```

```python
def tikz_code(deformation: jnp.ndarray) -> None:
    code = SurfaceCode(int(jnp.sqrt(deformation.shape[0])))
    parity_info = code.deformation_parity_info(deformation)
    print(code.show_latex_code(parity_info, deformation))
```

# Distance 3 (PML)

```python
# Random search
tikz_code(jnp.array([0, 1, 5, 4, 2, 3, 5, 1, 2]))
```

```python
# Hill climb
tikz_code(jnp.array([4, 2, 0, 2, 2, 2, 0, 2, 3]))
```

```python
# RL
tikz_code(jnp.array([0, 3, 0, 0, 5, 4, 0, 5, 5]))
```

```python
# Thermodynamic limit
tikz_code(jnp.array([2, 2, 1, 3, 5, 4, 2, 1, 0]))
```
